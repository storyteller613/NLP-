### Assumptions Multiple Linear Regression: 
#### 1. The mean of Yis linearly related to X. 
#### 2. Errors are normally distributed. 
#### 3. Errors have equal variances. 
#### 4. Errors are independent. 

```sas
libname statdata "/folders/myfolders/ECST131"; 
libname library "/folders/myfolders/ECST131";
```

### Task: Run a regression of PctBodyFat2 on the variables Age, Weight, Height, Neck, Chest, Abdomen, Hip, Thigh, Knee, Ankle, Biceps, Forearm, and Wrist.

### Results: The multiple regression model explains almost 75 percent of the variation in the PctBodyFat2 variable, versus only about 8.13 percent explained by the simple linear regression model with only Weight. 	 

```sas
title 'Regression of PctBodyFat2 on All Predictors';
proc reg data=statdata.bodyfat2;
   model PctBodyFat2 = Age Weight Height
         Neck Chest Abdomen Hip Thigh
         Knee Ankle Biceps Forearm Wrist;
run;
quit;
title;
```

```sas
title 'Regression of PctBodyFat2 on All Predictors';
proc reg data=statdata.bodyfat2;
   model PctBodyFat2 = Age;
run;
quit;
title;
```

### Task: To simplify the model, rerun the model from step 1, but eliminate the variable with the highest p-value. 

### Result: The variable with the highest p-value is Knee, and it is removed. The p-value for the model did not change out to four decimal places. The R^2 showed essentially no change. The adjusted R^2 increased from .7348 to .7359. When an adjusted R^2 increases by removing a variable from the models, it strongly implies that the removed variable was not necessary.

```sas
proc reg data=statdata.bodyfat2;
   model PctBodyFat2 = Age Weight Height
         Neck Chest Abdomen Hip Thigh
         Ankle Biceps Forearm Wrist;
   title 'Remove Knee';
run;
quit;
title;
```
