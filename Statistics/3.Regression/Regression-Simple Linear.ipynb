### Task: Using the BodyFat2 data set, produce a simple linear regression model with confidence and prediction intervals around the observations. Then produce predicted values. Produce confidence and prediction intervals around the observations.

### Result: The F value is 150.03 and the p-value is <.0001. You would reject the null hypothesis of no relationship. From the parameter estimates table, the predicted value of PctBodyFat2 = -12.05158 + 0.17439 * Weight. The R2 value of 0.3751 can be interpreted to mean that 37.51% of the variability in PctBodyFat2 can be explained by Weight.

```sas
libname statdata "/folders/myfolders/ECST131"; 
libname library "/folders/myfolders/ECST131";
```

```sas
title "Regression of % Body Fat on Weight";
proc reg data=statdata.bodyfat2;
   model PctBodyFat2=Weight;
run;
quit;
title;

title "Regression of % Body Fat on Weight";
proc reg data=statdata.bodyfat2;
   model PctBodyFat2=Weight / clm cli;
run;
quit;
title;
```

### Task: Produce predicted values for PctBodyFat2 when Weight is 125, 150, 175, 200, and 225.

### Result: The predicted values for PctBodyFat2 when Weight is 125, 150, 175, 200, and 225, are 9.7470, 14.1067, 18.4664, 22.8261, and 27.1859.

```sas
title "Regression of % Body Fat on Weight";
proc reg data=statdata.bodyfat2 outest=estimates;
   model PctBodyFat2=Weight;
run;

data toscore;
   input Weight @@;
   datalines;
125 150 175 200 225
;
run;

proc score data=toscore score=estimates
     out=scored type=parms;
   var Weight;
run;

title "Predicted % Body Fat from Weight 125 150 175 200 225";
proc print data=scored;
   
run;
title;
```
