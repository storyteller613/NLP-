### Task: Assess the model from the final forward stepwise selection of predictors for the Statdata.BodyFat2 data set by plotting residuals by the four regressors, plotting the residuals by the predicted values, and a normal quantile-quantile plot.

### Result: It does not appear that the data violates the assumption of constant variance based on the residual plots. Based on the Q-Q plot, the normality assumption seems to be met.

```sas
libname statdata "/folders/myfolders/ECST131"; 
libname library "/folders/myfolders/ECST131";
```

```sas
ods graphics / imagemap=on;

title 'FORWARD Model - Plots of Diagnostic Statistics';
proc reg data=statdata.bodyfat2 
         plots(only)=(QQ RESIDUALBYPREDICTED RESIDUALS);
   FORWARD: model PctBodyFat2 = Abdomen Weight Wrist Forearm;
   id Case;   
run;
quit;
title;
```
