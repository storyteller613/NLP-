### Task: The insurance company wants to characterize the relationship between the region in which a car was manufactured and its safety rating. Fit a simple logistic regression model with Unsafe as the outcome variable and Region as the predictor variable.

### Results: All three global tests (Likelihood Ratio, Score, and Wald) lead to the conclusion of not rejecting the null hypothesis. The equation is: Logit(Unsafe) = -0.2876 â€“ 0.8329 * (Region='N America'). The odds ratio for Region says that the odds of having a below average safety rating are 56.5% lower for cars produced in North America, compared with those produced in Asia. The confidence interval includes 1, indicating that that the odds ratio is not statistically significant.

```sas
title1 'LOGISTIC MODEL (1):Unsafe=Region';
proc logistic data=statdata.safety plots(only)=(effect);
   class Region (param=ref ref='Asia');
   model Unsafe(event='1')=Region;
run;

title;
```
